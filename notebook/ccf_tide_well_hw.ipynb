```python
import sys
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib as mpl
import matplotlib.ft2font as ft
import matplotlib.font_manager as font_manager
mpl.rcParams['figure.dpi'] = 300
import matplotlib.ft2font as ft
import matplotlib.font_manager as font_manager
plt.rcParams['axes.labelweight'] = 'bold'
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['axes.titleweight'] = 'bold'
plt.rcParams['axes.titlesize'] = 14
ccfpath = r'C:\Users\Martina\Google Drive\tide_induced_gw_level\spreadsheets\ccf'
```

```python
twpath = os.path.join( ccfpath, 'tide_well_ccf_grouped_tw_pos.csv') 
df_fall = pd.read_csv(twpath)
fig, ax = plt.subplots(1, 1,  figsize=(12, 5))
# make grouped boxplot and save it in a variable
#bp = sns.boxplot(y='r', x='Monitoring well and salinity structure', 
bp = sns.boxplot(y='r', x='Monitoring well',                  
                 data=df_fall, 
                 color= "turquoise", 
                  hue='group')
#,width=0.4 
# make grouped stripplot and save it in a variable
#bp = sns.stripplot(y='r', x='Monitoring well and salinity structure', 
bp = sns.stripplot(y='r', x='Monitoring well', 
                   data=df_fall, 
                   jitter=True,
                   dodge=True, 
                   marker='o', 
                   alpha=0.5,
                   hue='group',
                   color='grey')
# get legend information from the plot object
handles, labels = bp.get_legend_handles_labels()
# specify just one legend
l = plt.legend(handles[0:1], labels[0:1])
```
