```python
import pandas as pd
from pandas import read_csv
from pandas import datetime
from matplotlib import pyplot
import matplotlib.pyplot as plt
from scipy import interpolate

def parser(x):
    return datetime.strptime('201'+x, '%m/%d/%Y')

series = pd.read_csv(r'C:\Users\Martina\Google Drive\tide_induced_gw_level\spreadsheets\stages\s27_t_2018_2020.csv', header=0) 
series
```

```python
format = '%m/%d/%Y %H:%M'
series['datetime'] = pd.to_datetime(series['date'] , format=format)
series = series.set_index(pd.DatetimeIndex(series['datetime']))
series1 = series['tide'].resample(rule='H').mean().interpolate(method='cubic', order=3)
series1 = series1.reset_index()
series1['tide'] = (series1['tide']-1.52) # from ft NGVD to meters NAVD
series1.to_csv(r'C:\Users\Martina\PROJECTS\Jacobs model\test.csv') 
print(series1)
#plt.plot(series1)
```
