```python
import sys
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import read_csv
import datetime
%matplotlib inline
import statsmodels
import statsmodels.tsa.stattools as ts
from statsmodels.tsa.stattools import acf, pacf
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import scipy.stats as ss
#import pingouin as pg
#from pingouin import corr
import scipy.signal as sig
ccfpath = r'C:\Users\Martina\Google Drive\tide_induced_gw_level\spreadsheets\no_rain_ts'
```

```python
tspath = os.path.join( ccfpath, 'all_miami_beach2020.csv') 
df = pd.read_csv(tspath)
df= df.loc[0:230]
print(len(df))
well = df['NI']*0.3048
trend_well = well.rolling(12,center=True).mean() #12-hour mean trend
de_well = well - trend_well
np.savetxt('WELL.csv',de_well)
tide = df['s27t']*0.3048
trend_tide =tide.rolling(12,center=True).mean()
de_tide = tide - trend_tide
np.savetxt('test.csv',de_tide)
dw = de_well[~np.isnan(de_well)]
dt = de_tide[~np.isnan(de_tide)]
dw = np.array(dw)
dt = np.array(dt)

r = sig.correlate(dw - np.mean(dw), dt - np.mean(dt), method='direct') / (np.std(dw) * np.std(dt) * len(dw))
halfr = int(len(r)/2)+1
r1 = r[:halfr]
lag = r.argmax() - (len(dt) - 1)

print('Tide and well')
rmax = np.max(r)
rmin = np.min(r)
n = len(dw)
print('n=',n)
if abs(rmax)>abs(rmin):
    lag = r.argmax() - (len(dt) - 1)
    print('rmax=',rmax)
    t = (rmax*((n-2)**0.5))/((1-rmax**2)**0.5)
    p = ss.t.sf(abs(t), df=n-2)
else:
    lag = r.argmin() - (len(dt) - 1)
    print('rmin=',rmin)
    t = (rmin*(n-2)**0.5)/(1-rmin**2)**0.5
    p = ss.t.sf(abs(t), df=n-2)
print('rmax=',rmax)
print(r.argmax() - (len(dt) - 1))
print('p-value= ',p)    
print('lag= ',lag)  

plt.figure(figsize=(5,3))
#plt.plot(well)
#plt.plot(trend_well)
plt.plot(de_well)

plt.figure(figsize=(5,3))
plt.plot(tide)
plt.plot(trend_tide)
plt.plot(de_tide)
```
